Statistiques de Hopkins :
La statistique de Hopkins, est une statistique qui donne une valeur qui indique la tendance de cluster, en d'autres termes : dans quelle mesure les données peuvent être regroupées.

Si la valeur est comprise entre {0.01, ...,0.3}, les données sont régulièrement espacées.

Si la valeur est d'environ 0,5, c'est aléatoire.

Si la valeur est comprise entre {0,7, ..., 0,99}, elle a une forte tendance à se regrouper.

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

Algorithme K-means
Clustering hiérarchique
 construire le modèle en utilisant à la fois les méthodes et regrouper les pays identifier les pays dans le besoin

L'algorithme Kmeans est un algorithme itératif qui tente de partitionner l'ensemble de données en K sous-groupes (clusters) distincts et non superposés prédéfinis, où chaque point de données appartient à un seul groupe.

4.1 Métriques pour choisir la valeur de K
Le principal défi de cet algorithme est de trouver la valeur optimale de k ou nombre de clusters. Il existe deux approches courantes qui aident à trouver k :

Méthode du coude
Analyse des silhouettes
4.1.1 Méthode du coude
La méthode du coude nous donne une idée de ce qu'un bon nombre k de clusters serait basé sur la somme de la distance au carré (SSD) entre les points de données et les centroïdes des clusters qui leur sont attribués. Nous sélectionnons k à l'endroit où SSE commence à s'aplatir et à former un coude.

Utilisons KMeans() de sklearn pour former des clusters de 2,3,4 et ainsi de suite jusqu'à 11 et calculons SSD et traçons le nombre de clusters et SSD et voyons où le coude est formé.

```python
ssd = []
for k in range(2, 11):
    kmean = KMeans(n_clusters = k).fit(data_scaled)
    ssd.append([k, kmean.inertia_])
    
temp = pd.DataFrame(ssd)
ax = plt.axes()
ax.plot(temp[0], temp[1]) 
ax.axvline(3, ls='dotted',color='red') 
plt.xlabel('Number of clusters')
plt.ylabel('SSD')
plt.show()
```

SSD s'aplatit et forme un coude à 3 indiquant que 3 est la valeur optimale de k.


Analyse de silhouette:
Le score de silhouette est une mesure de la similitude d'un objet avec son propre cluster (cohésion) par rapport à d'autres clusters (séparation).

Utilisons KMeans() de sklearn pour former des clusters de 2,3,4 et ainsi de suite jusqu'à 11 et calculons silhouette_score et traçons le nombre de clusters par rapport à silhouette_score

```python
# Silhouette score

from sklearn.metrics import silhouette_score
silhouette_scores_list = []
for k in range(2, 11):
    kmean = KMeans(n_clusters = k).fit(data_scaled) # intialise kmeans
    silhouette_avg = silhouette_score(data_scaled, kmean.labels_) # silhouette score
    silhouette_scores_list.append([k, silhouette_avg])
    print("For k_clusters={0}, the silhouette score is {1:2f}".format(k, silhouette_avg))
    
temp = pd.DataFrame(silhouette_scores_list)    
ax = plt.axes()
ax.plot(temp[0], temp[1])
ax.axvline(3, ls='dotted',color='green') # elbow formed as 3
ax.axvline(4, ls='dotted',color='blue') # elbow formed as 3
ax.axvline(5, ls='dotted',color='maroon') # elbow formed as 3
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```


Le score de silhouette est maximum lorsque k vaut 2 soit 0,34
2 est un nombre très inférieur de clusters et les pays au sein des 2 clusters peuvent être très différents.
Regardons donc le prochain score optimal de silhouette.

3,4,5 semblent avoir de bons scores de silhouette. À mesure que k augmente, le score de silhouette diminue et, par conséquent, ceux-ci auront un score de sihouette nettement inférieur à celui de k = 2.
Bien que la courbe du coude indique que 3 est le nombre optimal et que le score de silhouette de 3 semble être le meilleur, utilisons l'algorithme K-means pour k = 3,4,5 et voyons quelle valeur de k nous donne un meilleur profilage de cluster

```python
# Function for all steps of Kmean Clustering; Call with K=3,4,5
def K_means_model(k):
    kmean = KMeans(n_clusters = k, random_state = 50+k)
    kmean.fit(data_scaled)
    country_df_kmean = data.copy() # copy the actual data into a new dataframe to explain the cluster profiling
    label  = pd.DataFrame(kmean.labels_, columns= ['k_means_cluster_label'])
    country_df_kmean = pd.concat([country_df_kmean, label], axis =1) # assign the countries with the cluster labels.
    print("Number of countries in each cluster(k=%s):" %k)
    print(country_df_kmean.k_means_cluster_label.value_counts())# shows how many countries are in each cluster
    return(country_df_kmean) # returns clustered labelled dataset for further analysis
```

```python
# Created Models are available globally to access inside cluster profiling functions
k_3_model = K_means_model(3) # K means model with 3 clusters
k_4_model = K_means_model(4) # K means model with 4 clusters
k_5_model = K_means_model(5) # K means model with 5 clusters
```

```python
# Subset the data for Tunisia and sdg6_water
df_tunisia_water = dft_imp.loc[(dft_imp['id'] == 'TUN') & (dft_imp['Year'] <= 2022), ['Year', 'sdg6_water']]

# Fit a linear regression model to the data
X = df_tunisia_water['Year'].values.reshape(-1, 1)
y = df_tunisia_water['sdg6_water'].values.reshape(-1, 1)
model = LinearRegression().fit(X, y)

# Predict the values for the years 2023-2030
X_future = np.arange(2023, 2031).reshape(-1, 1)
y_pred = model.predict(X_future)

# Create a DataFrame with the predicted values
df_pred_water = pd.DataFrame({'Year': X_future.flatten(), 'sdg6_water_pred': y_pred.flatten()})
# Subset the data for Tunisia and sdg8_unemp
df_tunisia_unemp = dft_imp.loc[(dft_imp['id'] == 'TUN') & (dft_imp['Year'] <= 2022), ['Year', 'sdg8_unemp']]

# Fit a linear regression model to the data
X = df_tunisia_unemp['Year'].values.reshape(-1, 1)
y = df_tunisia_unemp['sdg8_unemp'].values.reshape(-1, 1)
model = LinearRegression().fit(X, y)

# Predict the values for the years 2023-2030
X_future = np.arange(2023, 2031).reshape(-1, 1)
y_pred = model.predict(X_future)

# Create a DataFrame with the predicted values
df_pred_unemp = pd.DataFrame({'Year': X_future.flatten(), 'sdg8_unemp_pred': y_pred.flatten()})

# Merge the two DataFrames on the 'Year' column
df_pred = pd.merge(df_pred_water, df_pred_unemp, on='Year')

# Plot the predicted values
plt.plot(df_tunisia_water['Year'], df_tunisia_water['sdg6_water'], label='Actual')
plt.plot(df_pred['Year'], df_pred['sdg6_water_pred'], label='Predicted')
plt.plot(df_tunisia_unemp['Year'], df_tunisia_unemp['sdg8_unemp'], label='Actual')
plt.plot(df_pred['Year'], df_pred['sdg8_unemp_pred'], label='Predicted')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Predicted Values for SDG6 Water and SDG8 Unemployment in Tunisia')
plt.legend()
plt.show()
```
